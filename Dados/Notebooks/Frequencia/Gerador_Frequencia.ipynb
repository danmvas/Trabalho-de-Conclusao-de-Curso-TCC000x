```python
import random
from datetime import datetime,timedelta
import pandas as pd
import gc
import os
```

```python
def gen_dates(start,n):
    dates = []
    start = datetime.strptime(start, '%Y-%m-%d')
    dates.append(datetime.strftime(start,'%Y-%m-%d'))
    d = 1
    while len(dates) < n :
        date = start + timedelta(days=d)
        dates.append(datetime.strftime(date,'%Y-%m-%d'))
        d+=1
        if d % 5 == 0:
            d+=2
    return dates
```

```python
def mock_frequencia(n_alunos,n_alunos_turma,n_anos,n_dias_letivos):

    if not os.path.exists('dados_frequencia'):
        os.makedirs('dados_frequencia') 

    id_turma = 1
    for ano in range(n_anos):
        frequencias = []
        dates = gen_dates('202'+str(ano)+'-01-15',n_dias_letivos,)
        print('gerando dados do ano de 202'+str(ano))
        for id_aluno in range(1,n_alunos+1):
            for f in range(len(dates)):
                frequencia = {}
                frequencia['data'] = dates[f]
                frequencia['co_aluno'] = id_aluno
                frequencia['co_turma'] = id_turma
                prob_presente = random.randint(0,10)
                if prob_presente < 8:
                    frequencia['presenca'] = 1
                else:
                    frequencia['presenca'] = 0
                frequencias.append(frequencia)

            if (id_aluno) % n_alunos_turma == 0:
                id_turma += 1
                print('202'+str(ano)+' turma '+str(id_turma))
        print('escrevendo dados: ','202'+str(ano))
        df = pd.DataFrame(frequencias)
        df.to_csv('dados_frequencia/frequencia202'+str(ano)+'.csv',index=False)
        del df
        del frequencias
        gc.collect()
```

```python
n_alunos = 10
frequencia = mock_frequencia(n_alunos,50,3,200)
```
